```python
import pandas as pd
import json
```

```python
df = pd.read_json('a3cleanedonly2015.json')
df.head()
```

```python
df.drop(["FIELD1", 'Unnamed: 0', "UID"], 1)
```

```python
df.count()
```

```python
mi_prop = df.value_counts("Mental_illness").to_frame("counts").reset_index()
mi_prop.to_json("mh_counts.json", "records")
```

```python
states_df = df[(df['Mental_illness'] == True) & (df["State"].isin(["MI", "IL", "OH", "IN", "WI", "MN"]))].value_counts("State").to_frame("counts").reset_index()
states_df.head()
```

```python
states_df.to_json("state_counts.json", orient="records")
```

```python
age_df = df[df["Mental_illness"] == True]
age_df.dropna(subset = ['Age'])
age_df["Age"].to_frame("counts").to_json("age_dist.json", orient = "records")
```

```python
age_counts = age_df.value_counts('Age').to_frame("counts").reset_index(level = 0)
age_counts.to_json("age_dist.json", orient = "records")
```
